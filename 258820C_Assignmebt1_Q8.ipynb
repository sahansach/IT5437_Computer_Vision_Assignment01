Q8 An image corrupted with salt and pepper noise.

(a) Apply Gaussian smoothing.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load noisy image
img = cv.imread('/Users/sahansach/Documents/MSC/Semester 3/Computer Vision/Assignment 01/emma2.png', cv.IMREAD_REDUCED_GRAYSCALE_2)

# Apply Gaussian smoothing
gaussian_smoothed = cv.GaussianBlur(img, (5,5), 1.0)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Original (Salt & Pepper)")
plt.imshow(img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Gaussian Smoothing")
plt.imshow(gaussian_smoothed, cmap='gray')
plt.axis("off")
plt.show()
```

(b) Apply median filtering.

```python
# Apply median filtering
median_filtered = cv.medianBlur(img, 5)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Original (Salt & Pepper)")
plt.imshow(img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Median Filtering")
plt.imshow(median_filtered, cmap='gray')
plt.axis("off")
plt.show()
```
