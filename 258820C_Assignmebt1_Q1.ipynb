```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

f = cv.imread('/Users/sahansach/Documents/MSC/Semester 3/Computer Vision/Assignment 01/runway.png', cv.IMREAD_REDUCED_GRAYSCALE_2)
assert f is not None

gamma = 0.5
f_gamma = np.array([(i / 255) ** gamma * 255 for i in np.arange(0,256)]).astype(np.uint8)

fig, axs = plt.subplots(1, 3, figsize=(10, 5))
axs[0].imshow(f, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')  
axs[1].imshow(f_gamma[f], cmap='gray')
axs[1].set_title('Gamma Corrected Image')
axs[1].axis('off')
axs[2].plot(f_gamma, color='blue')
axs[2].set_title('Gamma Correction Curve')
axs[2].set_xlabel('Input Intensity')
axs[2].set_ylabel('Output Intensity')
axs[2].grid()
plt.tight_layout()
plt.show()  

```
