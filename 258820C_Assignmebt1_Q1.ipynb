(a) Apply the intensity transform to the runway image
Gamma = 0.5

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

#load the image in grayscale mode
f = cv.imread('/Users/sahansach/Documents/MSC/Semester 3/Computer Vision/Assignment 01/runway.png', cv.IMREAD_REDUCED_GRAYSCALE_2)
assert f is not None

# Apply gamma correction with gamma = 0.5
gamma = 0.5

# Create a lookup table for gamma correction
f_gamma = np.array([(i / 255) ** gamma * 255 for i in np.arange(0,256)]).astype(np.uint8)

# Apply the gamma correction to the image
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# Display the original image, gamma corrected image, and the gamma correction curve
axs[0].imshow(f, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')  
axs[1].imshow(f_gamma[f], cmap='gray')
axs[1].set_title('Gamma Corrected Image')
axs[1].axis('off')
axs[2].plot(f_gamma, color='blue')
axs[2].set_title('Gamma Correction Curve')
axs[2].set_xlabel('Input Intensity')
axs[2].set_ylabel('Output Intensity')
axs[2].grid()
plt.tight_layout()
plt.show()  

```

(c) Contrast Stretching (linear piecewise transformation)

```python
# Parameters
r1, r2 = 0.2, 0.8

# Apply inear contrast stretching
def contrast_stretching(r):
    if r < r1:
        return 0
    elif r > r2:
        return 1
    else:
        return (r - r1) / (r2 - r1)

# Vectorize the function for efficiency
contrast_func = np.vectorize(contrast_stretching)
img_stretched = contrast_func(f / 255.0)  # Normalize to [0,1] before stretching

# Rescale back to [0,255] for display
img_stretched = (img_stretched * 255).astype(np.uint8)

# Show original and transformed images
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(f, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Contrast Stretched Image")
plt.imshow(img_stretched, cmap='gray')
plt.axis("off")

plt.show()
```
