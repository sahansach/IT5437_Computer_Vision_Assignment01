The Convolution Theorem states that the Fourier Transform of a convolution of two functions is equal to the point-wise product of their individual Fourier Transforms.
In the context of 2D image and filter:
> Transformation: First transform both the image f (x, y) and the filter h(x, y) into the frequency domain using the Discrete Fourier Transform (DFT), resulting in 𝐹(u,𝑣) and 𝐻(u,𝑣).
> Point-wise Multiplication: The complex spatial operation of convolution (*) becomes simple multiplication in the frequency domain: G(u,v) = F(u,v)⋅H(u,v)

> Inverse Transformation: To get the filtered image 𝑔(𝑥,𝑦) back in the spatial domain, you apply the Inverse DFT to 𝐺(𝑢,𝑣).


(b) Qualitative Effects of Filters
Averaging (box) filter
Spatial: Smooths by averaging neighbors, blurs edges.
Frequency: Acts as a low‑pass filter, attenuates high frequencies. Produces sinc‑like frequency response with side lobes → can cause ringing.


Gaussian filter
Spatial: Smooths gently, preserves overall structure.
Frequency: Also a low‑pass filter, but with a Gaussian frequency response (no sharp cutoff). Smooth attenuation of high frequencies.


Laplacian filter
Spatial: Highlights edges by detecting rapid intensity changes.
Frequency: High‑pass filter, amplifies high frequencies (edges, noise), suppresses low frequencies (smooth regions).

(c) Why Gaussian Avoids Ringing
Ideal low‑pass filter has a sharp cutoff in frequency domain → corresponds to sinc kernel in spatial domain. The sinc has oscillations, which cause ringing artifacts near edges.
Gaussian filter has a smooth frequency response → no abrupt cutoff, so no oscillations in spatial domain. Hence, Gaussian avoids ringing.

(d) Best Filter for High‑Frequency Noise
Noise is typically high‑frequency.
Averaging filter reduces noise but introduces artifacts (ringing, edge blur).
Laplacian filter amplifies noise (bad choice).
Gaussian filter is most suitable:
Spatial domain: smooths noise while preserving structure.
Frequency domain: attenuates high frequencies gradually, avoiding ringing.
